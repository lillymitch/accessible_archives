##Abstract
This project explores whether machine learning models can predict the time period/era of an artwork using only image data. Using artwork images streamed from a Hugging Face dataset, we developed and evaluated several image classification models, including a Logistic Regression baseline model, a custom Convolutional Neural Network (CNN), and a transfer learning model using ResNet18. The goal of this project was to determine whether visual features extracted from artwork images could be used to classify artworks into distinct artistic categories, including Impressionism, Cubism, Baroque, Realism, and Expressionism.

The project involved streaming and preprocessing image data, filtering selected classes, creating training, validation, and testing splits, and training multiple models to compare performance. Because of computational and time limitations, we processed a limited subset of the available dataset while still maintaining representation across all selected classes. Several preprocessing and augmentation techniques were applied, including image resizing, normalization, random cropping, and horizontal flipping. Our experiments demonstrated that transfer learning with ResNet18 provided the strongest performance because pretrained convolutional features captured meaningful visual patterns that simpler models struggled to identify. In addition to evaluating overall accuracy, we analyzed confusion matrices, optimizer performance, and common classification errors.

This project demonstrates how machine learning can assist museums, researchers, and digital archives in organizing large art collections while also highlighting the limitations and ethical concerns associated with automated artwork classification systems.


#Introduction
Digital museum collections now contain thousands of digitized artworks, creating new opportunities for museums, researchers, and educators to analyze and organize art using computational methods. However, manually categorizing artworks by style, historical period, or artistic characteristics remains difficult and time-consuming, especially when metadata is incomplete or inconsistent. Accurately identifying the era of an artwork is important because it helps museums improve collection management, supports art historical research, and increases accessibility for public audiences. As digital archives continue to grow, automated image classification systems have become increasingly valuable tools for managing and interpreting fine art collections.

Recent research has explored the use of machine learning and convolutional neural networks (CNNs) for fine art analysis and classification. Cetinic, Lipic, and Grgic (2018) investigated the use of fine-tuned CNNs for multiple art-related classification tasks, including style, genre, artist, and time-period classification. Their study demonstrated that transfer learning with pre-trained CNN models significantly improved classification accuracy compared to traditional handcrafted feature methods. Similarly, Cetinic, Lipic, and Grgic (2020) examined how CNNs could learn stylistic principles from art history by predicting visual characteristics associated with Heinrich Wölfflin’s theories of artistic style. Their findings showed that deep learning models could identify meaningful stylistic patterns within artworks and support new approaches to exploring digital art collections. More recently, Imran et al. (2023) proposed a hybrid deep learning framework that combined deep and shallow neural networks for painting style classification. Their approach achieved high classification accuracy by analyzing image patches and using transfer learning techniques, demonstrating the effectiveness of increasingly sophisticated neural network architectures for fine art recognition tasks.

Building on this prior research, the purpose of this project was to investigate whether a machine learning model could accurately predict the era of an artwork using only the image itself. Because artwork classification depends heavily on visual characteristics such as color, texture, brush strokes, and composition, image classification techniques were selected as the primary approach. This project specifically explored three different modeling methods: a logistic regression baseline model, a custom convolutional neural network (CNN), and a transfer learning model using ResNet18. By comparing these approaches, the project aimed to evaluate how increasing model complexity influenced classification performance and determine which method was most effective for artwork era prediction.

#Dataset and Data Preparation
Although our original proposal planned to use the Metropolitan Museum of Art Open Access dataset, we ultimately encountered accessibility and retrieval issues when attempting to consistently access large numbers of artwork images through the museum API. Because of these limitations, we transitioned to using a publicly available artwork image dataset hosted on Hugging Face and changed our predicted feature to the artwork’s era.

The Hugging Face dataset provided a more reliable and streamlined method for accessing large collections of labeled artwork images. Instead of downloading all images locally at once, the dataset was streamed directly during processing, which reduced storage requirements and simplified data loading.

To keep the project computationally manageable within the available project timeline, we imposed a maximum number of examples to process from the dataset. This allowed us to train and evaluate multiple models efficiently while still maintaining a sufficiently diverse set of artwork images.
The original dataset contained a large number of artistic styles and categories. To simplify the classification problem and focus on visually distinct styles, we narrowed the dataset to the following five classes: Impressionism, Cubism, Baroque, Realism, and Expressionism. These categories were selected because they contain recognizable stylistic differences that could potentially be learned by image classification models.

After filtering the selected classes, the dataset was divided into separate training, validation, and testing splits. The training set was used for model learning, the validation set was used for tuning and monitoring model performance, and the testing set was reserved for final evaluation. Creating separate splits helped prevent overfitting and ensured that model performance was evaluated on previously unseen images.


####Image Preprocessing
Several preprocessing and augmentation techniques were applied to prepare the images for training. All images were transformed using PyTorch and torchvision preprocessing utilities which included image resizing, tensor conversion, and image normalization. The images were resized to a consistent input size so they could be processed uniformly by the neural network architectures. We then normalized the images, helping stabilize training and improve convergence during optimization. Normalization values were selected to match ImageNet preprocessing standards, which was particularly important for the transfer learning model. Random cropping was applied as a form of data augmentation. This technique randomly selects and and crops different portions of an image before resizing the crop to the final input size. It also helps decrease model sensitivity to exact image positioning and encourages it to learn more general visual features rather than memorizing specific image layouts. Random horizontal flipping was used to expose the model to additional visual variations without requiring additional data collection. This helps improve generalization performance and reduces overfitting. Several challenges emerged during data preparation: some artistic styles contained significantly more examples than others, the quality of artwork images varied, many artistic styles shared overlapping visual characteristics, and computational limitations required restricting the total number of processed examples. Despite these limitations, the final dataset provided a strong foundation for training and comparing multiple models for classification.


####Train/validation/test split
We divided the dataset into 70% training data, 15% validation data, and 15% testing data. The training set was used to optimize model weights, the validation set was used for hyperparameter turning and monitoring performance during training, and the test set was used for final evaluation.


# Our Approach
To evaluate different approaches to image classification, we implemented and compared three separate model architectures. Our first model was a Logistic Regression (LR) classifier. This served as a baseline model that allowed us to compare more advanced neural network approaches against a simpler linear method. Although LR is computationally efficient and easy to interpret, it struggles with image data because it cannot naturally capture spatial relationships or hierarchical visual patterns. The model flattened the image tensor into a one dimensional vector before classification.

Our second model was a custom CNN implemented in Pytorch. CNNs are well-suited for image classification tasks because convolutional layers can automatically learn local visual features such as edges, shapes, textures, and color patterns. Our CNN architecture included: convolutional layers, ReLU activation functions, pooling layers, and fully connected classification layers. The CNN extracted visual features directly from the artwork images and learned hierarchical representations during training.

Our final model used transfer learning with ResNet18. Transfer learning allows a model pretrained on a large dataset to be adapted to a new task. ResNet18 was originally trained on ImageNet, which contains millions of labeled images across many categories. Instead of training an entirely new network from scratch, we reused pretrained convolutional layers and replaced the final classification layer to match our artwork medium categories. This approach provided several advantages such as faster training, better feature extraction, improved performance on limited datasets, and reduced overfitting. Transfer learning was particularly effective because the pretrained model had already learned useful low-level image features such as edges, textures, and shapes.


###Training Procedure and Optimization
All of our models were trained using Cross Entropy Loss. We experimented with multiple optimizers to compare training performance and convergence behavior. Optimizer comparison was important because different optimization algorithms can significantly affect convergence speed, training stability, and final accuracy.

Training was performed using PyTorch with GPU acceleration when CUDA was available. The device selection process used:


```
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

GPU acceleration significantly reduced training time for the CNN and ResNet18 models.
The training loop includes forward propagation, loss computation, back propagation, weight updates, and validation accuracy evaluation. Training and validation accuracy are monitored across epochs to evaluate model learning progress.



#Results and Performance Analysis:

The LR baseline model achieved relatively low classification accuracy. Example results include:


*   Training accuracy: approximately 22-28%
*   Validation accuracy: approx 24-27%
*   Test accuracy: approx 23%

These results demonstrate the limitations of linear models for image classification tasks. The confusion matrix also shows that the LR model struggles to distinguish between several classes and frequently predicts the same categories repeatedly.
Example confusion matrix:


```
[ 0, 0, 5, 1, 7],
[ 0, 0, 5, 0, 10],
[ 0, 0, 9, 1, 8],
[ 0, 0, 3, 2, 9],
[ 0, 0, 5, 2, 6]
```


This confusion matrix suggests that the model fails to learn strong distinguishing features for several classes.

Our CNN performed substantially better than LR because convolutional layers were able to learn spatial image features. The CNN demonstrated improved ability to identify textures and visual patterns associated with different artistic eras. However, the CNN still faces challenges due to limited dataset size, class imbalance, and similar visual appearance between some eras.
The ResNet18 transfer learning model achieved the strongest overall performance. Transfer learning significantly improves our feature extraction because the pretrained convolutional layers have already learned generalized visual representations from ImageNet. Compared to the baseline and the CNN, ResNet18 demonstrated better validation accuracy, more stable training, faster convergence, and improved generalization. This result supports the idea that pretrained computer vision models are highly effective for specialized image classification tasks with limited datasets.
